Q1. What is multiprocessing in python? Why is it useful?

  multiprocessing:- multiprocessing is a module that allows you to create and manage processes,
                   which are independent units of execution that run concurrently. It provides an interface 
                   for spawning processes, passing data between them, and synchronizing their execution.
                   
                   
        1. Parallel Execution: By using multiprocessing, you can leverage multiple CPU cores to execute tasks in parallel.
           This is particularly beneficial for computationally intensive or time-consuming tasks, as it can significantly 
           reduce the overall execution time.
           
           
        2. Improved Performance: When executing tasks concurrently, multiprocessing allows you to take advantage of the
           available CPU resources more efficiently. This can lead to improved performance and increased throughput
           or certain types of applications.  
           
           
        3.Resource Isolation: Each process created with multiprocessing has its own memory space, which provides
          isolation between processes. This means that if one process encounters an error or crashes, it won't affect 
          other processes. Additionally, memory is not shared by default, preventing unintended side effects due to
          concurrent access to shared data.
          
          
        4. Enhanced Responsiveness: By offloading intensive or time-consuming tasks to separate processes,
           the main process can remain responsive and continue to handle other tasks or user interactions without
           being blocked. This is particularly relevant in applications that require a high degree of interactivity
           or real-time responsiveness.
           
           
        5.Distributed Computing: Multiprocessing can be used to distribute tasks across multiple machines,
          enabling distributed computing. This allows you to scale your application across a cluster of computers,
          harnessing their combined computational power.

Q4. What is a multiprocessing pool in python? Why is it used?

 Python, a multiprocessing pool is a convenient abstraction provided by the multiprocessing module.
         It allows you to create a pool of worker processes that can execute tasks in parallel.
         The pool manages the creation and management of the worker processes, simplifying the 
         process of distributing tasks and collecting results.
    
    
    
    
      To use a multiprocessing pool, you first need to create a Pool object. The Pool constructor 
      takes an argument that specifies the number of worker processes to create. Once you have 
      created a Pool object, you can use it to execute tasks by calling the apply_async() or map() methods.
      

       apply_async() method takes a function and a tuple of arguments as its arguments.
          The map() method takes a function and an iterable of arguments as its arguments.
          In both cases, the Pool object will execute the function in a separate worker process
          and return a Future object. The Future object can be used to get the result of the function call,
          or to check if the function has finished executing.



```python
from multiprocessing import Pool

def my_function(x):
  return x * x

if __name__ == "__main__":
  pool = Pool(4) 
  results = pool.map(my_function, range(10)) 
  for result in results:
    print(result) 
```

Q6. Write a python program to create 4 processes, each process should print a different number using the
    multiprocessing module in python.

```python
import multiprocessing

def print_number(number):
    print(f"Process {number}: This is number {number}.")

if __name__ == '__main__':
    processes = []

    for i in range(1, 5):
        process = multiprocessing.Process(target=print_number, args=(i,))
        processes.append(process)
        process.start()

    for process in processes:
        process.join()

    print("All processes have finished.")
```
